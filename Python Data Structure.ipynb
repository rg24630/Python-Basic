1. What are data structures, and why are they important?
- Data structures are a way of efficiently arranging and storing information on a computer. They are essential for programmers because they enable efficient data management and manipulation, which improves performance and makes code maintenance simpler.

2.  Explain the difference between mutable and immutable data types with examples?
-In programming, mutable data types can be modified after they are created, while immutable data types cannot. Lists, dictionaries, and sets are mutable, whereas strings and tuples are immutable. This means you can change the elements within a mutable object, but with immutable objects, any modification results in a new object being created.
Example:
Mutable Data Types:List, Dictionaries, Sets
Immutable Data Types:Strings, Tuples.

3. What are the main differences between lists and tuples in Python?
-Both lists and tuples can hold collections of data in Python, but they differ in terms of memory usage, performance, and mutability. Tuples are immutable, but lists are mutable, enabling changes. Depending on whether you need to change the data or give performance and memory efficiency priority, you can choose between them.

7. How do tuples ensure data integrity in Python?
-Python tuples' immutability is the main way they guarantee data integrity. Because of this feature, a tuple's contents cannot be altered, added to, or removed once it has been created. Data integrity is strongly guaranteed by its immutability.

8. What is a hash table, and how does it relate to dictionaries in Python?
--Python dictionaries and hash tables are related in that the latter is a high-level abstraction that is based on the former. When you use Python dictionaries, Python manages the hashing, collision resolution, and storage details behind the scenes, so you don't have to deal with the underlying hash table implementation directly. As a result, developers can map keys to values using a practical and effective data structure without having to deal with the hassles of implementing hash tables.

9.Can lists contain different data types in Python?

10. Explain why strings are immutable in Python?
-- Python strings are immutable, which means that once they are formed, their contents cannot be altered. Any operation that appears to modify a string, such as concatenation or replacement, actually results in the creation of a new string object, while the original string remains unchanged.

11. What advantages do dictionaries offer over lists for certain tasks?
--Lists are best suited for sequential processing, sorted collections, and straightforward data storage, but dictionaries perform best in situations demanding quick lookups, key-value connections, and intricate data mappings.

12. Describe a scenario where using a tuple would be preferable over a list?
--The tuples are unchangeable. As a result, their main purpose is to store data that isn't updated often. If you do not want data to change, you can save it in a tuple using any operation. Use tuples if you want the information in your collection to be read-only, unchangeable, and consistent at all times.

13. How do sets handle duplicate values in Python?
--Python set objects are designed to store only unique elements. This means that when duplicate values are added to a set, they are automatically ignored and only one instance of the value is stored. Sets inherently enforce uniqueness.

14. How does the “in” keyword work differently for lists and dictionaries?

15. Can you modify the elements of a tuple? Explain why or why not?
--No, Python does not allow you to directly alter a tuple's elements. Since tuples are immutable, their contents cannot be altered once they are created. One of the fundamental properties of tuples is their immutability.


16. What is a nested dictionary, and give an example of its use case?
--A dictionary that has at least one value that is also a dictionary is called a nested dictionary. It is a method of using a dictionary structure to represent structured or hierarchical data. This enables effective organization and retrieval of complex knowledge.

17. Describe the time complexity of accessing elements in a dictionary?
--No matter the size of the dictionary, accessing its elements (also called a hash map or associative array) usually takes the same amount of time, with an average temporal complexity of O(1). This is due to the fact that dictionaries store and retrieve data internally using hash tables, and hash table lookups are typically quite quick.

18. In what situations are lists preferred over dictionaries?
--Dictionaries for fast configurations, caches, or data lookups. Lists are better suited for sequence operations and ordered collections, like keeping a stack or queue.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
--Dictionary data structures are significant value structures that are optimized for quick storage and key-based recovery. This is why dictionaries are an unordered collection. Simply put, Python dictionaries are not meant to be in order, which is why they are un-ordered.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
---Lists are retrieved using an index, whereas dictionaries are retrieved using a key. This means that in a dictionary, you use the associated key to find the corresponding value, whereas in a list, you use the element's integer position (beginning at 0) to retrieve it.
